```python
import cv2
import os
import numpy as np

def rotate_image(image_path, output_folder, angle):
    # Read the image
    img = cv2.imread(image_path)

    # Get the rotation matrix
    rows, cols, _ = img.shape
    M = cv2.getRotationMatrix2D((cols / 2, rows / 2), angle, 1)

    # Apply the rotation
    rotated_img = cv2.warpAffine(img, M, (cols, rows))

    # Get the filename without extension
    filename_without_extension = os.path.splitext(os.path.basename(image_path))[0]

    # Save the rotated image to the output folder
    output_path = os.path.join(output_folder, f"{filename_without_extension}_rotated.jpg")
    cv2.imwrite(output_path, rotated_img)

def process_images(input_folder, output_folder, rotation_angle):
    # Create the output folder if it doesn't exist
    if not os.path.exists(output_folder):
        os.makedirs(output_folder)

    # List all files in the input folder
    image_files = [f for f in os.listdir(input_folder) if f.lower().endswith(('.png', '.jpg', '.jpeg'))]

    # Process each image
    for image_file in image_files:
        image_path = os.path.join(input_folder, image_file)
        rotate_image(image_path, output_folder, rotation_angle)

if __name__ == "__main__":
    # Set the input folder, output folder, and rotation angle
    input_folder = "data/horse_photos/ShetlandPony"
    output_folder = "Rotated-ShetlandPony"
    rotation_angle = 20  # Adjust this value as needed

    # Process the images
    process_images(input_folder, output_folder, rotation_angle)
```
